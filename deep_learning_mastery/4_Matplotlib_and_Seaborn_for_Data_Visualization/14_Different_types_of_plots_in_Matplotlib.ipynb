# Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
points = np.array([4, 7, 2, 10, 13])
```

```python
plt.plot(points)
plt.show()
```

```python
a2 = np.array([1, 2, 3, 4, 5, 6])
plt.plot(a2, linestyle = 'dashed')
plt.show()
```

# Bar Chart

```python
x = np.array(["X", "Y", "Z", "W"])
y = np.array([25, 30, 27, 22])
```

```python
plt.bar(x, y)
plt.show()
```

# Scatter Plot

```python
x1 = np.array([20, 15, 11, 7, 8, 10])
y1 = np.array([40, 20, 5, 10, 11, 1])
```

```python
plt.scatter(x1, y1)
plt.show()
```

# Pie Chart

```python
p = np.array([60, 80, 30])
```

```python
plt.pie(p)
```
